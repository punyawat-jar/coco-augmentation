```python

os.chdir('/home/punyawat/BUU-LSPINE_400/python/coco-augmentation/')
import os
import shutil
import json
from tqdm import tqdm
from pycocotools.coco import COCO


def move_images_by_id(choose_categories_id):
    dir = ['AP','LA']
    
    for i in dir:
        images = []
        src = f'../../class_dataset/split/{i}/train'
        dst = f'./source/{i}'
        
        src_json = f'../../class_dataset/{i}.json'
        dst_json = f'./source/{i}.json'

        cleanDst(dst)
        os.remove(dst_json)
        shutil.copy(src_json, dst_json)
        os.makedirs(f'./source/{i}', exist_ok= True)
        
        coco = COCO(dst_json)
        with open(dst_json, 'r') as file:
            data = json.load(file)
            
    
        for item in data['annotations']:
            if item['category_id'] in choose_categories_id:
                images.append(coco.loadImgs(item['image_id'])[0]['file_name'])

        for image in tqdm(images, desc = 'Processing Images'):
            src_file = f'{src}/{image}'
            dst_file = f'{dst}/{image}'
            
            if os.path.exists(src_file):
                shutil.copyfile(src_file, dst_file)
        
def cleanDst(dst_images):
    if os.path.exists(dst_images):
        shutil.rmtree(dst_images)


```

```python
from pycocotools.coco import COCO
import albumentations as A
import cv2

import numpy as np
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt 
coco = COCO(dst_json)
```

```python
coco.loadAnns([1])
```

```python
coco.loadImgs([1])
```

```python
image = cv2.imread('./source/AP/0001-F-037Y0.jpg')
```

```python
def augmentingImage(image, masks, augmentation):

    augmented = augmentation(image=image, masks=masks)
    
    image_augmented = augmented['image']
    mask_augmented = augmented['masks']
    
    return image_augmented, mask_augmented

def read_corr_coco(coco, key):
    item = []
    for _, value in coco.anns.items():
        if value['image_id'] == key:
            item.append(value)
            
    return item

def getMask(image, corr):
    masks = []
    anns = []
    for polygon in corr:
        anns.append(polygon['category_id'])
        mask_img = Image.new('L', (image.shape[1], image.shape[0]), 0)
        draw = ImageDraw.Draw(mask_img)
        draw.polygon(polygon['segmentation'][0], fill=1)
        mask = np.array(mask_img)
        masks.append(mask)
    return masks, anns


def mask2polygon(mask):
    contours, _ = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    segmentation = []

    for contour in contours:
        contour = contour.flatten().tolist()
        if len(contour) > 5:
            segmentation.append(contour)

    return segmentation
def processMasklist(coco_mask):
    flag = 0
    for i in range(0, len(coco_mask) - 1):
        if coco_mask[i] == coco_mask[i+1]:
            flag = i
            break
        
    if flag != 0:
        for i in reversed(range(i+1, len(coco_mask))):
            coco_mask.pop(i)

    for i in reversed(range(len(coco_mask))):
        if coco_mask[i] == []:
            coco_mask.pop(i)
    return coco_mask


```

```python
corr = read_corr_coco(coco, 1)
            
masks, anns = getMask(image, corr)
```

```python
x = [1,2,3]
y = ['a','b','c']

for i, j in x,y:
    print(i,j)
```

```python
augmentation = A.Compose([
        A.RandomCrop(always_apply=True, p=1.0, height=640, width=640)
    ])

image_augmented, mask_augmented = augmentingImage(image, masks, augmentation)
```

```python
coco_mask = []
coco_anns = []
for mask in mask_augmented:
    new_coor = mask2polygon(mask)
    coco_mask.append(new_coor)


for i, mask in enumerate(coco_mask):
    if mask != []:
       coco_anns.append(anns[i])
```
