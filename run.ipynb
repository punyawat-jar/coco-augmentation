# Prepare images by calss

```python
import os
os.chdir('/home/punyawat/BUU-LSPINE_400/python/coco-augmentation/')
import shutil
import json
from tqdm import tqdm
from pycocotools.coco import COCO


def move_images_by_id(choose_categories_id):
    dir = ['AP','LA']
    
    for i in dir:
        images = []
        src = f'../../class_dataset/split/{i}/train'
        dst = f'./source/{i}'
        
        src_json = f'../../class_dataset/{i}.json'
        dst_json = f'./source/{i}.json'

        cleanDst(dst)
        os.remove(dst_json)
        shutil.copy(src_json, dst_json)
        os.makedirs(f'./source/{i}', exist_ok= True)
        
        coco = COCO(dst_json)
        with open(dst_json, 'r') as file:
            data = json.load(file)
            
    
        for item in data['annotations']:
            if item['category_id'] in choose_categories_id:
                images.append(coco.loadImgs(item['image_id'])[0]['file_name'])

        for image in tqdm(images, desc = 'Processing Images'):
            src_file = f'{src}/{image}'
            dst_file = f'{dst}/{image}'
            
            if os.path.exists(src_file):
                shutil.copyfile(src_file, dst_file)
        
def cleanDst(dst_images):
    if os.path.exists(dst_images):
        shutil.rmtree(dst_images)


```
